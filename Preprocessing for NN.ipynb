## Detecting Outliers using box plot

```python
def draw_univariate_plot(dataset, rows, cols, plot_type):
    column_names=dataset.columns.values
    number_of_column=len(column_names)
    fig, axarr=plt.subplots(rows,cols, figsize=(20,10))

    counter=0
    
    for i in range(rows):
        for j in range(cols):

            if counter >= number_of_column:
                axarr[i][j].set_visible(False) 
                continue

            if column_names[counter]=='target':
                break

            if 'violin' in plot_type:
                sns.violinplot(x='target', y=column_names[counter],data=dataset, ax=axarr[i][j])
            elif 'box'in plot_type : # You can add other plot types here
                sns.boxplot(x=None,y=column_names[counter],data=dataset,ax=axarr[i][j],width=0.5)

            axarr[i][j].set_title(column_names[counter])
            counter += 1
            
            if counter==(number_of_column-1,):
                break
```

```python
z = df.drop(['Sex', 'Category'], axis=1)
draw_univariate_plot(dataset=z, rows=3, cols=5, plot_type="box")
plt.subplots_adjust(hspace=0.2, wspace=0.4) 
plt.show()
```

## Removing rows based on IQR Conditions(Optional)

```python
numeric_df = df.select_dtypes(include=['number'])

for col in numeric_df.columns:
    Q1 = numeric_df[col].quantile(0.25)
    Q3 = numeric_df[col].quantile(0.75)
    
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    numeric_df = numeric_df[(numeric_df[col] >= lower_bound) & (numeric_df[col] <= upper_bound)]

df_cleaned = pd.concat([numeric_df, df.select_dtypes(exclude=['number'])], axis=1)
df_cleaned.drop(['Sex', 'Category'], axis=1, inplace=True)
print(numeric_df.columns)
print(numeric_df.shape)
```

```python
draw_univariate_plot(dataset=df_cleaned, rows=3, cols=5, plot_type="box")
plt.subplots_adjust(hspace=0.2, wspace=0.4) 
plt.show()
```

# Data Preprocessing/Cleaning

## Filling Missing values bases on skewness of the data

```python
df.isna().sum()
```

```python
df.select_dtypes(include=['int64', 'float64']).skew()
```

```python
df['ALB'] = df.ALB.fillna(df.ALB.mean())
df['ALP'] = df.ALP.fillna(df.ALP.median())
df['ALT'] = df.ALT.fillna(df.ALT.median())
df['CHOL'] = df.CHOL.fillna(df.CHOL.mean())
df['PROT'] = df.PROT.fillna(df.PROT.median())
```

## Separating features and target variable

```python
X = df.drop('Category', axis=1)
Y = pd.DataFrame(df['Category'], index=X.index, columns=['Category'])

print(X, Y)
```

## Converting Gender Feature to binary variable

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
gender_transform = le.fit_transform(X['Sex']).astype(int)
X['Sex'] = gender_transform

gender_classes = le.classes_
print(le.classes_) # F=0, M=1
print(X['Sex'].sample(5))

label_mapping = {label: index for index, label in zip(le.transform(gender_classes), gender_classes)}
print("Label Mapping Dictionary:", label_mapping)
```

## Converting Categorical Target Classes to Numerical Target Classes using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
class_transform = le.fit_transform(Y['Category']).astype(int)
Y['Category'] = class_transform
print(le.classes_)
print(Y['Category'].sample(5))

label_mapping = {label: index for index, label in zip(le.transform(le.classes_), le.classes_)}

print("Label Mapping Dictionary:", label_mapping)
```
